```python


# Initialize the Boto3 client for Amazon Lex
lex_client = boto3.client('lexv2-models', region_name='us-east-1')

# Define the bot name and slot type name
bot_name = 'SearchBot'
slot_type_name = 'Object'

# Define the values for the slot type
slot_values = [
    "Tree", "Mountain", "Beach", "Sunset", "Flower", "Dog", "Cat", "Bird", "Waterfall",
    "Sky", "Cloud", "Cityscape", "Building", "Architecture", "Car", "Bike", "Hiking",
    "Camping", "Wildlife", "Lake", "River", "Boat", "Ship", "Bridge", "Sunflower", "Field",
    "Forest", "Snow", "Ice", "Firework"
]



# Create or update the slot type
response = lex_client.create_slot_type(
    botId='FSDIULYSZH',
    localeId='en_US',
    botVersion='DRAFT',
    slotTypeName=slot_type_name,
    description='Description of your slot type',
    slotTypeValues=[{'sampleValue': {'value': value}} for value in slot_values],
    valueSelectionSetting= { 'resolutionStrategy':'TopResolution'}
)

# Print the response
print(response)
```
